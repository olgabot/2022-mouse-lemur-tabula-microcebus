## Add sex and age

```python
sex_mapper = {
    "Martine": "female",
    "Stumpy": "female",
    "Bernard": "male",
    "Antoine": "male",
}

age_mapper = {
    "Martine": "12y",
    "Stumpy": "10y",
    "Bernard": "9y",
    "Antoine": "12y",
}

lemur_tenx.obs["sex"] = lemur_tenx.obs.individual.map(sex_mapper)
lemur_tenx.obs["age"] = lemur_tenx.obs.individual.map(age_mapper)
```

## Add lemur number to individual name

```python
lemur_to_number = dict(zip(sorted(lemur_tenx.obs.individual.unique()), range(1, 5)))
lemur_to_number
```

```python
lemur_tenx.obs['individual_v2'] = lemur_tenx.obs.individual.map(lambda x: f"lemur_{lemur_to_number[x]}_{x}")
lemur_tenx.obs['individual_v2'].value_counts()
```

# Add unified cell type groups

## Use minimal obs columns

```python
obs_minimal = lemur_tenx.obs.rename(columns={'batch_SS2': None,
 'cell_barcode_10x': 'cell_barcode',
 'cell_name': 'cell_id_v2',
 'cell_name_old': 'cell_id_v1',
 'cell_ontology_class': 'cell_ontology_class',
 'channel': 'channel',
 'compartment': None,
 'free_annotation': 'free_annotation',
 'individual': None,
 'individual_v2': 'individual',
 'method': None,
 'nCount_RNA': 'n_counts',
 'nFeature_RNA': 'n_genes',
 'possibly_contaminated_barcode_10x': None,
 'sequencing_run_10x': 'sequencing_run',
 'species': 'species',
 'species_latin': 'species_latin',
 'subtissue': 'subtissue',
 'tissue': 'tissue',
 'tissue__cell_ontology_class': None,
 'tissue__free_annotation': None})
obs_minimal = obs_minimal.loc[:, obs_minimal.columns.notnull()]
print(obs_minimal.shape)
obs_minimal.head()
```

## Blood

```python
import unified_annotations

blood_grouping = unified_annotations.get_celltype_converter('Blood', ("Lemur", "Tabula Microcebus", "free_annotation_update"))
print(blood_grouping.shape)
blood_grouping.head()
```

## Lung

```python
%load_ext autoreload
%autoreload 2
```

```python
lung_grouping = unified_annotations.get_celltype_converter('Lung', ("Lemur", "Tabula Microcebus", "free_annotation_update"))
print(lung_grouping.shape)
lung_grouping.head()
```

## Muscle

```python
muscle_grouping = unified_annotations.get_celltype_converter('Muscle', ("Lemur", "Tabula Microcebus", "free_annotation_update"))
print(muscle_grouping.shape)
muscle_grouping.head()
```

## Concatenate

```python
grouping_concatenated = pd.concat([blood_grouping, muscle_grouping, lung_grouping])
grouping_concatenated = grouping_concatenated[["narrow_group", "broad_group", "compartment_group"]]
grouping_concatenated = grouping_concatenated.loc[~grouping_concatenated.index.duplicated()]
print(grouping_concatenated.shape)
grouping_concatenated.head()
```

## Mege grouping with minimal obs

```python
new_obs = obs_minimal.merge(grouping_concatenated, left_on='free_annotation', right_index=True, how='left')
print(new_obs.shape)
new_obs.head()
```

## Make new object with new metadata

```python
lemur_tenx_new_obs = lemur_tenx.copy()
lemur_tenx_new_obs.obs = new_obs
```
