## Concatenate counts

```python

counts = pd.concat([mouse1_counts, mouse2_counts])
describe(counts)
```

```python
is_transgene = counts.columns.map(lambda x: x.endswith('TRANSGENE') or x.startswith('ERCC-')).values
sum(is_transgene)

# 
```

```python
def check_is_transgene(x):
    if x.startswith('ERCC-'):
        return True
    elif x.endswith('TRANSGENE'):
        return True
    else:
        return False
```

```python
non_transgenes = [x for x in counts.columns if not check_is_transgene(x)]
```

```python
len(non_transgenes)
```

```python
counts_no_transgenes = counts.loc[:, non_transgenes]
counts_no_transgenes
```

## Make columns capitalized, not all caps

```python
counts_no_transgenes.columns = counts_no_transgenes.columns.str.capitalize()
counts_no_transgenes.head()
```

## Concatenate metadata

```python
metadata = pd.concat([mouse1_metadata, mouse2_metadata])
describe(metadata)
```

# Add unified annotations

```python
import unified_annotations

blood_grouping = unified_annotations.get_celltype_converter(
    "Blood", ("Mouse", "CE independent dataset", "free_annotation")
) 
blood_grouping = blood_grouping.drop('tissue', axis=1)
print(blood_grouping.shape)
blood_grouping.head()
```

```python
metadata_unified = metadata.merge(
    blood_grouping, left_on="free_annotation", right_index=True
)
describe(metadata_unified)
```

```python
sorted(metadata_unified.columns)
```

## Add individual_v2 column

```python
dict.fromkeys(metadata_unified.individual.unique())
```

```python
indidivual_renamer = {"Mouse_1": "mouse_24_ce1", "Mouse_2": "mouse_25_ce2"}
metadata_unified["individual_v2"] = metadata_unified.individual.map(indidivual_renamer)
metadata_unified["individual_v2"].value_counts()
```

## Rename metadata columns

```python
metadata_unified_renamed = metadata_unified.rename(
    columns={
        "broad_group": "broad_group",
        "channel": "channel",
        "compartment": None,
        "compartment_group": "compartment_group",
        "free_annotation": "free_annotation",
        "individual": None,
        "individual_v2": "individual",
        "method": None,
        "nGene": "n_genes",
        "nUMI": "n_counts",
        "narrow_group": "narrow_group",
        "orig.ident": None,
        "percent.ribo": None,
        "processed": None,
        "region": None,
        "species": "species",
        "tissue": "tissue",
    }
)
metadata_unified_renamed["species_latin"] = "Mus musculus"
metadata_unified_renamed = metadata_unified_renamed.loc[
    :, metadata_unified_renamed.columns.notnull()
]
print(metadata_unified_renamed.shape)
metadata_unified_renamed.head()
```

```python
adata = anndata.AnnData(X=counts_no_transgenes.loc[metadata_unified_renamed.index], obs=metadata_unified_renamed)
adata
```
