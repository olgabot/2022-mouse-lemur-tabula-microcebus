## Pick a region

```python
utar_id = 'NC-033660.1-87624599-87626899---848-0'
```

```python
# utar_homology['utar_id'] = utar_homology['region_id'].str.upper()
this_utar_protein = utar_protein.query('utar_id == @utar_id')
print(this_utar_protein.shape)
this_utar_protein = this_utar_protein.set_index('read_id')
this_utar_protein.columns = 'protein__' +  this_utar_protein.columns
this_utar_protein.head()
```

```python
this_utar_rna = utar_rna.query('utar == @utar_id')
print(this_utar_rna.shape)
this_utar_rna = this_utar_rna.set_index('target_name')
this_utar_rna.columns = 'rna__' + this_utar_rna.columns
print(this_utar_rna.shape)
this_utar_rna.head()
```

```python
utar_predictions.loc[utar_id]
```
