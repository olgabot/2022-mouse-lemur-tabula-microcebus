```python
# sigma = 20
# prob_false_positive_coding_frame = 1e-50
# ksize = 9

def compute_best_translation_length(prob_false_positive_coding_frame, sigma, ksize, bloomfilter):
    bf_collisions = khmer.calc_expected_collisions(bloomfilter) 
    numerator = math.log(prob_false_positive_coding_frame) + (ksize - 1) * math.log(bf_collisions)
    denominator = math.log(prob_false_positive_coding_frame) - math.log(sigma)
    print(f"numerator: {numerator}")
    print(f"denominator: {denominator}")
    result = numerator / denominator
    return result

compute_best_translation_length(prob_false_positive_coding_frame=1e-100, sigma=20, ksize=9, bloomfilter=index_mammals)
```
